## **Python Lists**

- A dynamic array, used to store multiple items is a single variable. 
- List items are:
    - Ordered (items have defined order, If we add new it will be placed at the end)
    - Changable 
    - Allows duplicate values 

- Fast indexing 
- But shifting is expensive

```python
mylist= [10, 20, 40, "Lion", True, 30, 90, 10, 40]  #Can be of different datatypes
print(mylist)
```

```python
##Length of list
print(len(mylist))
```

```python
print(type(mylist))
```

```python
##Indexing
print(mylist[0])
```

```python
print(mylist[1])
```

```python
##Negative indexing = reverse traversal 
print(mylist[-1])
```

- ### Traversing list

```python
##Index based traversing (for two pointers, binary search, etc..)
for i in range(len(mylist)):  
    print(mylist[i])
```

```python
##Value based 
for x in mylist:
    print(x)
```

```python
##Accessing by range of indexes
print(mylist[2:5])    # 2 included, 5 not included
```

```python
print(mylist[:4])  #From beginning, not includes 4
```

```python
print(mylist[3:])
```

- Changing list items

```python
mylist[2]=30
print(mylist)
```

```python
mylist[4:7]= False, 20, 80
print(mylist)
```

- Adding List Items

```python
##Adds at the end
mylist.append("Lakshmi")
print(mylist)
```

```python
#Inserting at a specific index
mylist.insert(4, 40)
print(mylist)
```
