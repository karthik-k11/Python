### **Sliding Window**

Sliding Window is a technique used to process a contiguous block of elements by moving a window over the array instead of recomputing values repeatedly.

(or)

Sliding Window is an optimization technique that reduces time complexity by maintaining a window of elements and updating results as the window moves.

**When to think sliding window ?**

Think of Sliding Window when:

- Problem involves subarray or substring

- Elements must be continuous

We need:

- Maximum / Minimum

- Sum

- Count

- Window size is fixed OR condition-based

**Problem**
Maximum sum subarray of size K

**Statement**
Given an array of integers and a number k,
find the maximum sum of any subarray of size k.
