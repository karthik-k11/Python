## **Tuples**
- A tuple is an ordered collection of elements that cannot be changed (immutable).
- We cannot add, remove and update items.
- Allows Duplicate
- Using Round brackets

```python
tuple1= ("Peacock", "Pegion", "Parrot")
print(tuple1)
```

- Single item

```python
#Using comma while storing a single item
tuple2 = ("Lion",)
print(type(tuple2))

#This is not a tuple this is a string
tuple3 = ("apple")
print(type(tuple3))
```

- Accessing Tuples

```python
tupleA= ("Pen", "Book", 1, False)
print(tupleA[1])
```

```python
#Negative indexing
print(tupleA[-1])
```

```python
#Range of indexes
print(tupleA[2:4])

print(tupleA[:4])

print(tupleA[1:])
```

```python
#Range of negative indexes
print(tupleA[-4:-2])
```

- Check if Item exists

```python
if "Pen" in tupleA:
    print("Yes, Pen is there")
else:
    print("It's not there")
```

```python
if "Zen" in tupleA:
    print("Yes, Pen is there")
else:
    print("It's not there")
```

### Update Tuples

- We cannot add, remove or modify the items in the tupels, as it is called as *immutable*
- But to update tuple, we can convert it to list, update the list and convert it back to tuples. 

```python
a= ("Chatgpt", "Gemini", "Grok") #This is a tuple

b= list(a)     #Converted to list

b[0]="Claude"    #Updated the item of index 0

a= tuple(b)     #Converted back to tuple
print(a)
```

- Add items 

```python
#using append()
c= list(a)
c.append("Blackbox")
a=tuple(c)
print(a)

print(type(a))
```

- Remove items

```python
#Remove() function
d= list(a)
d.remove("Blackbox")
a=tuple(d)
print(a)
```
