### **Sliding Window**

Sliding Window is a technique used to process a contiguous block of elements by moving a window over the array instead of recomputing values repeatedly.

(or)

Sliding Window is an optimization technique that reduces time complexity by maintaining a window of elements and updating results as the window moves.

**When to think sliding window ?**

Think of Sliding Window when:

- Problem involves subarray or substring

- Elements must be continuous

We need:

- Maximum / Minimum

- Sum

- Count

- Window size is fixed OR condition-based

**Problem**
Maximum sum subarray of size K

**Statement**
Given an array of integers and a number k,
find the maximum sum of any subarray of size k.

### **Brute Force Approach**

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

max_sum = float('-inf')

for i in range(len(arr) - k + 1):
    window_sum = 0
    for j in range(i, i + k):
        window_sum += arr[j]
    max_sum = max(max_sum, window_sum)

print(max_sum)
```

TC:
- Outer Loop: O(n)
- Inner Loop: O(k)

So, TC is:
O(nk)

### **Sliding Window Approach**

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

#Compute first window sum
window_sum = sum(arr[:k])
max_sum = window_sum

#Slide the window
for i in range(k, len(arr)):
    window_sum += arr[i]        
    window_sum -= arr[i - k]    
    max_sum = max(max_sum, window_sum)

print(max_sum)
```

Sliding windows works only for contiguous segments.

**TC:** 

O(n)

Because Only one loop over array.

**SC:**

O(1)

Only variables used.
