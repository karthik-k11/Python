## **Python Lists**

- A dynamic array, used to store multiple items is a single variable. 
- List items are:
    - Ordered (items have defined order, If we add new it will be placed at the end)
    - Changable 
    - Allows duplicate values 

- Fast indexing 
- But shifting is expensive

```python
mylist= [10, 20, 40, 30, 90, 10, 40]
print(mylist)
```

```python
##Indexing
print(mylist[0])
```

```python
print(mylist[1])
```

```python
##Negative indexing = reverse traversal 
print(mylist[-1])
```

- ### Traversing list

```python
##Index based traversing (for two pointers, binary search, etc..)
for i in range(len(mylist)):  
    print(mylist[i])
```

```python
##Value based 
for x in mylist:
    print(x)
```
