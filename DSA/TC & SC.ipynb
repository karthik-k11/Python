## **Time and Space Complexity**
- DSA is not about getting the answer, it's about getting the answer **Fast** and **Less memory**

### **Time Complexity**
- If the input becomes very big, Will my program survive ?.
- Measure growth, not seconds.

**Big O notation that should be remembered**
| Name           | Big-O       | Meaning      |
|----------------|-------------|--------------|
| Constant       | O(1)        | Super fast   |
| Logarithmic    | O(log n)    | Very fast    |
| Linear         | O(n)        | OK           |
| Linear Log     | O(n log n)  | Acceptable   |
| Quadratic      | O(n²)       | Slow         |
| Exponential    | O(2ⁿ)       | Very slow    |
 

**Examples**

```python
#Example 1
import array as arr

a= arr.array('i',[1,2,3,4,5])

def print_all(a):
    for x in a:
        print(x)

print_all(a)
```

So Here, if the number of elements in the array is 1000 there will be 1000 prints

Hence the Complexity is O(n).
That is linear.

```python
#Now, print pairs
def print_pairs(a):
    for i in a:
        for j in a:
            print(i,j)
print_pairs(a)
```

In the above cell, If number of elements = 100, It would be 10,000 prints

So, Complexity is O(n²). That is Quadratic

### **Space Complexity**
- How much extra memory the program will use.

```python
def copy_arr(a):
    new=[]
    for i in a:
        new.append(i)
        print(new)
copy_arr(a)
```

- Here, It uses extra list of size n.
- So, the SC is  O(n).

**What will be the TC if I insert an element a the beginning of an array ?**

```python
a= arr.array('i', [20, 30, 40, 50, 60])
a.insert(0,10)
print(a)
```

- The TC will be O(n)
- Because, when I insert an element at the beginning, All the elements will shift to right.
- n number of elements will be n shifts

**What if I insert at the end ?**

```python
a.pop(5)
print(a)
```

- Now, the TC will be O(1), because it is constant and does not change or modify the index of any other elements. 

## Important Rule

| Operation                     | Time Complexity |
|--------------------------------|-----------------|
| Access by index                | O(1)            |
| Traverse array                 | O(n)            |
| Insert at end (amortized)      | O(1)            |
| Insert at beginning / middle   | O(n)            |
| Delete from beginning / middle | O(n)            |


**What is Amortized TC ?**

Amortized time complexity is the average time taken per operation over a sequence of operations, where some operations may be expensive, but most are cheap.

Even though a single operation can take O(n) (like array resizing), the overall average cost per operation remains O(1).
