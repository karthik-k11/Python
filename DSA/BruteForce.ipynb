### **Brute Force Thinking**

- Solving a problem by trying all possible options without worrying about the efficiency first. 
- It is

       - Simple

       - Straight forward

       - Easy to explain

       - Often slow

       - But correct.

Let's Solve a problem using brute force method.

-  Given an array of integers, find the maximum sum of any pair

I/P: arr= [14, 12, 25, 41, 57]

```python
arr= [14, 12, 25, 41, 57]
max_sum= 0

n= len(arr)

for i in range(n):
    for j in range(i+1, n):
        max_sum= max(max_sum, arr[i]+arr[j])
print(max_sum)
```
